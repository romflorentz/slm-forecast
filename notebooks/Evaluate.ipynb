```python
import os
import pandas as pd

dfs = []
for subdir, _, files in os.walk("/Users/romflorentz/Documents/ClientRepos/slm-forecast/data/evaluations"):
    for file in files:
        df = pd.read_csv(os.path.join(subdir, file))
        df["Model"] = file.split(".")[0]
        dfs.append(df)
```

```python
pd.concat(dfs).pivot_table(index="Category", columns="Model", values=["R²"], aggfunc=["mean", "std"])
```

```python
pd.concat(dfs).pivot_table(index="Model", values=[ 'MAE', 'MSE', 'RMSE', 'R²', 'MAPE', 'sMAPE',], aggfunc=["mean", "std"]).T
```
